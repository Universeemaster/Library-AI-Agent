## Defining the model id

## Creating the agent
We need to create the agent using the properties we defined so far:

```python
client = APIClient(credentials=credentials, project_id=project_id, space_id=space_id)

# Create the chat model
def create_chat_model():
    chat_model = ChatWatsonx(
        model_id=model_id,
        url=credentials["url"],
        space_id=space_id,
        project_id=project_id,
        params=parameters,
        watsonx_client=client,
    )
    return chat_model
```

```python
from ibm_watsonx_ai.deployments import RuntimeContext

context = RuntimeContext(api_client=client)




def create_utility_agent_tool(tool_name, params, api_client, **kwargs):
    from langchain_core.tools import StructuredTool
    utility_agent_tool = Toolkit(
        api_client=api_client
    ).get_tool(tool_name)

    tool_description = utility_agent_tool.get("description")

    if (kwargs.get("tool_description")):
        tool_description = kwargs.get("tool_description")
    elif (utility_agent_tool.get("agent_description")):
        tool_description = utility_agent_tool.get("agent_description")
    
    tool_schema = utility_agent_tool.get("input_schema")
    if (tool_schema == None):
        tool_schema = {
            "type": "object",
            "additionalProperties": False,
            "$schema": "http://json-schema.org/draft-07/schema#",
            "properties": {
                "input": {
                    "description": "input for the tool",
                    "type": "string"
                }
            }
        }
    
    def run_tool(**tool_input):
        query = tool_input
        if (utility_agent_tool.get("input_schema") == None):
            query = tool_input.get("input")

        results = utility_agent_tool.run(
            input=query,
            config=params
        )
        
        return results.get("output")
    
    return StructuredTool(
        name=tool_name,
        description = tool_description,
        func=run_tool,
        args_schema=tool_schema
    )


def create_custom_tool(tool_name, tool_description, tool_code, tool_schema, tool_params):
    from langchain_core.tools import StructuredTool
    import ast

    def call_tool(**kwargs):
        tree = ast.parse(tool_code, mode="exec")
        custom_tool_functions = [ x for x in tree.body if isinstance(x, ast.FunctionDef) ]
        function_name = custom_tool_functions[0].name
        compiled_code = compile(tree, 'custom_tool', 'exec')
        namespace = tool_params if tool_params else {}
        exec(compiled_code, namespace)
        return namespace[function_name](**kwargs)
        
    tool = StructuredTool(
        name=tool_name,
        description = tool_description,
        func=call_tool,
        args_schema=tool_schema
    )
    return tool

def create_custom_tools():
    custom_tools = []


def create_tools(context):
    tools = []
    
    config = None
    tools.append(create_utility_agent_tool("GoogleSearch", config, client))
    config = {
    }
    tools.append(create_utility_agent_tool("DuckDuckGo", config, client))
    config = {
        "maxResults": 5
    }
    tools.append(create_utility_agent_tool("Wikipedia", config, client))
    config = {
    }
    tools.append(create_utility_agent_tool("WebCrawler", config, client))

    return tools
```

```python
def create_agent(context):
    # Initialize the agent
    chat_model = create_chat_model()
    tools = create_tools(context)

    memory = MemorySaver()
    instructions = """don't answer other questions unrelated to library and books.
You are \"Libris,\" an AI assistant designed to manage and enhance the operations of a modern library. Your primary goal is to provide exceptional service to both library patrons and staff. You should be knowledgeable, efficient, and friendly, with a tone that is helpful and authoritative without being overly formal.
Key Objectives:

Patron Assistance: Help library visitors find books, understand library policies, and use library resources.

Administrative Support: Assist library staff with tasks like managing inventory, tracking overdue books, and generating reports.

Information Retrieval: Access and synthesize data from the library's catalog, user database, and operational logs to answer questions accurately and in real-time.

Proactive Engagement: Suggest new books, highlight library events, and offer assistance without being prompted when a user's intent is clear.

Core Capabilities & Instructions
Your responses should be based on the following capabilities and instructions.

Book & Resource Search:

Find books, journals, and other materials by title, author, subject, or ISBN.

Provide the book's location (e.g., \"Non-fiction, Shelf 4B\"), availability (e.g., \"Available for checkout\"), and a brief summary.

Handle complex queries like \"Find science fiction books written in the 1990s\" or \"Suggest a book similar to Dune.\"

User Account Management:

Check a user's checkout history, including books currently checked out and their due dates.

Handle requests for renewals.

Inform users of any outstanding fines or overdue items.

Protect user privacy and only provide account-specific information when the user is properly authenticated (e.g., provides a library card number or a unique identifier).

Library Policies & Procedures:

Clearly explain policies regarding borrowing limits, fine structures, hours of operation, and special services (e.g., inter-library loans, meeting room reservations).

Provide step-by-step instructions for tasks like \"How do I reserve a computer?\" or \"What's the process for getting a new library card?\"

Operational Tasks (Staff Mode):

Log new book acquisitions, update the status of books (e.g., \"lost,\" \"damaged\"), and manage inventory.

Generate reports on popular genres, overdue items, or user statistics.

Create and send automated overdue notices to patrons.

Crucially, recognize the difference between a patron and a staff request. If a request seems administrative, ask for staff credentials before proceeding.

Data & Integration:

Your knowledge is dynamic and should be integrated with a live database of the library's catalog and user information.

If you don't have the information, state that clearly and offer to forward the request to a human librarian.

Examples of User Interactions
Patron Request:

User: \"Can you help me find a book?\"

Libris: \"Certainly! What is the title or author you're looking for?\"

Patron Request (Complex):

User: \"What are the three most popular fantasy books checked out in the last month?\"

Libris: \"Let me check the data. Based on recent checkouts, the three most popular fantasy books are: The Shadow's Edge, The Dragon's Ascent, and A Crown of Whispers. Would you like me to check their availability?\"

Staff Request:

User: \"I need a report on all overdue books from the last quarter.\"

Libris: \"To proceed with this administrative request, please provide your staff ID. Once verified, I can generate that report for you.\"

Failed Request:

User: \"When is the library open on Christmas Day?\"

Libris: \"I do not have specific holiday hours in my current data. I recommend checking the library's official website or calling the front desk for the most accurate information.\""""

    agent = create_react_agent(chat_model, tools=tools, checkpointer=memory, state_modifier=instructions)

    return agent
```

## Invoking the agent
Let us now use the created agent, pair it with the input, and generate the response to your question:


```python
agent = create_agent(context)

def convert_messages(messages):
    converted_messages = []
    for message in messages:
        if (message["role"] == "user"):
            converted_messages.append(HumanMessage(content=message["content"]))
        elif (message["role"] == "assistant"):
            converted_messages.append(AIMessage(content=message["content"]))
    return converted_messages

question = input("Question: ")

messages = [{
    "role": "user",
    "content": question
}]

generated_response = agent.invoke(
    { "messages": convert_messages(messages) },
    { "configurable": { "thread_id": "42" } }
)

print_full_response = False

if (print_full_response):
    print(generated_response)
else:
    result = generated_response["messages"][-1].content
    print(f"Agent: {result}")
```
